# Machine Learning (ML) Validation and Analysis

Please prepare the input file in .xlsx or .csv from the significance genes (DEG result). Some machine learning models are used, such as Random Forest, Neural Network, SVM, Naive Bayes, Logistic Regression, and XGBoost. The validation outputs are ROC, AUC, and Confusion Matrix, inspect the highest precision, accuracy, recall, F1, and MCC to choose the best ML model
