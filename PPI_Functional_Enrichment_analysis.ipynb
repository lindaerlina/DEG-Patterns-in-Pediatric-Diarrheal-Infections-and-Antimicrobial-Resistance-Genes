# Protein-Protein Interaction (PPI) and Functional Enrichment Analysis

Please input the list of genes (protein-coding only) into the Cytoscape software, the protein-protein interaction will consist of interaction values ranging from 0 to 1 (weakest to strongest interaction). Nodes are proteins and edges are protein-protein interactions
  

Protein-Protein Interaction (PPI) using Cytoscape software can be downloaded https://cytoscape.org/. Functional Enrichment for KEGG Pathway and Gene Ontology (GO) using ShinyGO https://bioinformatics.sdstate.edu/go/
